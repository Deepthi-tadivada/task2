```python
import pandas as pd
import pyodbc

conn = pyodbc.connect(
    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=DESKTOP-7A07QL7\\MSSQLSERVER1;'
    'DATABASE=ETL2;'  # Ensure the correct database name
    'UID=sa;'
    'PWD=1234;'
)

# Correct the table name and schema
query = "SELECT * FROM dbo.Customers"  # Use the correct table name

# Fetch data into a DataFrame
df = pd.read_sql(query, conn)

print(df)
conn.close()
```

```python
df.iloc[993]
```

```python
df['name']=df['name'].str.replace(r'\s*(Jr\.|Sr\.|II|III|MD|DDS|PhD)$',"",regex=True)
 
```

```python
df['name']=df['name'].str.replace(r'^(Mr\.|Mrs\.|Miss\.|Dr\.)\s*',"",regex=True)
```

```python
df.loc[993]
```

```python
df.columns
```

```python
df['name'] = df['name'].astype(str)

# Step 2: Split the 'name' column into 'first_name' and 'last_name'
df[['first_name', 'last_name']] = df['name'].str.split(' ', n=1, expand=True)

# Step 3: Handle missing last_name values (optional)
df['last_name'] = df['last_name'].fillna('Unknown')

# Display the resulting DataFrame
print(df)
```

```python
# Remove extra characters from the `phone` column
df['phone'] = df['phone'].str.replace(r'[^0-9]', '', regex=True)

# Display the updated DataFrame
print(df[['phone']].head())
```

```python
df['country_code'] = df['address'].str.extract(r'\b([A-Z]{2})\b(?!.*\b[A-Z]{2}\b)')

# Replace nulls with a placeholder for manual verification
df['country_code'].fillna('UNKNOWN', inplace=True)

```

```python
df.isnull().sum()
```

```python
df['email']=df['email'].fillna('NO EMAIL')
df['phone']=df['phone'].fillna('Unknown Phone')
```

```python
unified_model=pd.merge(df,df1,on='country_code',how='inner')
```

```python
tier_mapping = {
    "Gold": 2,
    "Silver": 1,
    "Bronze": 0
}

# Add Customer_Tier column to unified model DataFrame
unified_model['Customer_Tier'] = unified_model['loyalty_status'].map(tier_mapping)

# Display updated DataFrame
print(unified_model)
```
